```python
import pandas as pd
import spacy
from tqdm import tqdm
```

```python
# Load English spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv("test.csv", encoding="latin1")
```

```python
# We'll apply cleaning to these columns
text_columns = ["Question", "Response", "CorrectAnswer"]
```

```python
def clean_text_spacy(text):
    if pd.isna(text):
        return ""
    doc = nlp(text.lower())
    cleaned_tokens = [
        token.lemma_ for token in doc
        if not token.is_punct and not token.is_space
    ]
    return " ".join(cleaned_tokens)
```

```python
# Apply preprocessing to selected columns
for col in text_columns:
    tqdm.pandas(desc=f"Processing {col}")
    df[col] = df[col].progress_apply(clean_text_spacy)
```

```python
# Save cleaned dataset
df.to_csv("test_cleaned.csv", index=False)
```
